### **Importing the Libraries**

```python
import cv2
from PIL import Image
from IPython.display import display
from skimage import io
```

### **Loading the Image**

```python
img = io.imread("https://raw.githubusercontent.com/srijan-27/LGMVIP-DataScience/main/1596532888_ram-mandir-in-ayodhya-photos-10.jpg")
img = cv2.resize(img, (600, 500))
display(Image.fromarray(img))
```

### **Converting into Gray Scale**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_img))
```

### **Inverting the Image**

```python
inverted_img = cv2.bitwise_not(gray_img)
display(Image.fromarray(inverted_img))
```

### **Blurring the Image using Gaussian Blur**

```python
blurred = cv2.GaussianBlur(inverted_img, (5, 5), 0, 0)
display(Image.fromarray(blurred))
```

### **Inverting the Image to get the Final Sketch**

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=250.0)
display(Image.fromarray(pencil_sketch))
```

### **Comparing the Original Image with Pencil Sketch**

```python
display(Image.fromarray(img))
display(Image.fromarray(pencil_sketch))
```
